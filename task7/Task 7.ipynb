```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data1.csv')
df.plot()

plt.show()
```

```python
df
```

```python
import plotly
import plotly.graph_objs as go

x_label = 'Возраст'
y_label = 'Месячный платеж'
z_label = 'Доход'
marker_color = df['Благонадежный заемщик'].replace('Нет', 'red').replace('Да', 'blue')
marker_size = df['Опыт работы'][:df['Опыт работы'].size - 1]
marker_form = df['Пол'].replace('женский', 'square').replace('мужской', 'circle')[:df['Пол'].size - 1]

fig = go.Scatter3d(x = df[x_label],
                 y = df[y_label],
                 z = df[z_label],
                 marker=dict(
                     opacity=0.9,
                     color=marker_color,
                     size=marker_size,
                     symbol=marker_form,
                     reversescale=True,
                     colorscale='Blues'),
                 line=dict(width=0.02),
                 mode='markers')

mylayout = go.Layout(scene=dict(
                        xaxis=dict(title=x_label),
                        yaxis=dict(title=y_label),
                        zaxis=dict(title=z_label)))

plotly.offline.plot({
    'data': [fig],
    'layout': mylayout
}, auto_open=True)
```
